15. Program to implement a basic probabilistic prime test?

```python
import random

def fermat_test(n, k=5):
    if n <= 1:
        return False
    for _ in range(k):
        a = random.randint(2, n - 2)
        if pow(a, n - 1, n) != 1:
            return False
    return True

n = int(input("Enter number: "))

if fermat_test(n):
    print("Probably Prime")
else:
    print("Composite")
```

17. Program to implement the Extended Euclidean Algorithm?

```python
def extended_gcd(a, b):
    if b == 0:
        return a, 1, 0
    gcd, x1, y1 = extended_gcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return gcd, x, y

a = int(input("Enter a: "))
b = int(input("Enter b: "))

g, x, y = extended_gcd(a, b)
print("GCD:", g)
print("x:", x)
print("y:", y)
```

49. Program to identify subgroups of a given group?

```python
from itertools import combinations

def is_subgroup(subset, n):
    if 0 not in subset: return False
    for a in subset:
        for b in subset:
            if (a + b) % n not in subset: return False
    return True

n = int(input("Enter n for group (Zn, +): "))
elements = list(range(n))
subgroups = []

for i in range(1, n + 1):
    for combo in combinations(elements, i):
        if is_subgroup(combo, n):
            subgroups.append(combo)

print(f"Subgroups of Z_{n}: {subgroups}")
```

50. Program to verify field axioms programmatically?

```python
def verify_field_axioms(n):
    elements = list(range(n))
    # 1. Additive Abelian Group check
    # 2. Multiplicative Abelian Group (for non-zero elements)
    # 3. Distributivity
    non_zero = [e for e in elements if e != 0]

    has_inv = all(any((a * b) % n == 1 for b in non_zero) for a in non_zero)
    is_comm = all((a * b) % n == (b * a) % n for a in elements for b in elements)

    if has_inv and is_comm:
        print(f"Z_{n} satisfies field axioms (Commutativity & Multiplicative Inverses).")
    else:
        print(f"Z_{n} fails field axioms.")

n = int(input("Enter n to check axioms: "))
verify_field_axioms(n)
```

52. Program to compute Eulerâ€™s Phi Function using prime factorization?




```python
import math

def phi_gcd(n): # 51
    return len([x for x in range(1, n) if math.gcd(x, n) == 1])

def phi_prime_fact(n): # 52
    result = n
    p = 2
    temp = n
    while p * p <= temp:
        if temp % p == 0:
            while temp % p == 0:
                temp //= p
            result *= (1 - (1 / p))
        p += 1
    if temp > 1:
        result *= (1 - (1 / temp))
    return int(result)

n = int(input("Enter n to compute phi(n): "))
print(f"Phi (Counting GCD): {phi_gcd(n)}")
print(f"Phi (Prime Factorization): {phi_prime_fact(n)}")
```

60. Program to analyze the hardness of prime factorization for large integers?

```python
import time

def factorize(n):
    start = time.time()
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1: factors.append(n)
    return factors, time.time() - start

n = int(input("Enter a large composite number (e.g., product of two primes): "))
factors, duration = factorize(n)
print(f"Factors: {factors} | Time taken: {duration:.6f} seconds")
```

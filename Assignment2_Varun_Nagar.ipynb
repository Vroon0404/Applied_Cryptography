# **Problem 1**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def phi(n):
    result = n
    p = 2
    temp = n
    while p * p <= temp:
        if temp % p == 0:
            while temp % p == 0:
                temp //= p
            result -= result // p
        p += 1
    if temp > 1:
        result -= result // temp
    return result

def mod_add(a, b, m):
    return (a + b) % m

def mod_mul(a, b, m):
    return (a * b) % m

def mod_exp(a, b, m):
    return pow(a, b, m)

a = int(input("Enter first positive integer: "))
b = int(input("Enter second positive integer: "))

print("Prime check:", is_prime(a), is_prime(b))

g = gcd(a, b)
print("GCD:", g)

coprime = g == 1
print("Co-prime:", coprime)

if coprime:
    phi_a = phi(a)
    phi_b = phi(b)
    print("Phi(a):", phi_a)
    print("Phi(b):", phi_b)

    x = 2
    print("Fermat verification:", mod_exp(x, a - 1, a) if is_prime(a) else "Not applicable")
    print("Euler verification:", mod_exp(x, phi_a, a))

m = int(input("Enter modulus: "))
x = int(input("Enter x: "))
y = int(input("Enter y: "))

print("Modular Addition:", mod_add(x, y, m))
print("Modular Multiplication:", mod_mul(x, y, m))
print("Modular Exponentiation:", mod_exp(x, y, m))
```

# **Problem 2**

```python
def mod_add(a, b, m):
    return (a + b) % m

def mod_mul(a, b, m):
    return (a * b) % m

def is_closed(S, op):
    for a in S:
        for b in S:
            if op(a, b) not in S:
                return False
    return True

def has_identity(S, op):
    for e in S:
        if all(op(a, e) == a and op(e, a) == a for a in S):
            return e
    return None

def has_inverses(S, op, e):
    for a in S:
        if not any(op(a, b) == e and op(b, a) == e for b in S):
            return False
    return True

def is_associative(S, op):
    for a in S:
        for b in S:
            for c in S:
                if op(op(a, b), c) != op(a, op(b, c)):
                    return False
    return True

def is_distributive(S, add, mul):
    for a in S:
        for b in S:
            for c in S:
                if mul(a, add(b, c)) != add(mul(a, b), mul(a, c)):
                    return False
    return True

m = int(input("Enter modulus: "))
S = set(range(m))

add = lambda a, b: mod_add(a, b, m)
mul = lambda a, b: mod_mul(a, b, m)

add_id = has_identity(S, add)
mul_id = has_identity(S, mul)

add_group = (
    is_closed(S, add) and
    is_associative(S, add) and
    add_id is not None and
    has_inverses(S, add, add_id)
)

ring = (
    add_group and
    is_closed(S, mul) and
    is_associative(S, mul) and
    is_distributive(S, add, mul)
)

field = (
    ring and
    mul_id is not None and
    has_inverses(S - {0}, mul, mul_id)
)

print("\nResults:")
print("Additive Group:", add_group)
print("Ring:", ring)
print("Field:", field)

print("\nSample modular operations:")
print("Addition:", add(3, 5))
print("Multiplication:", mul(3, 5))
print("Exponentiation:", pow(3, 5, m))
```
